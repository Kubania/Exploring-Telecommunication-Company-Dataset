Merging the Telcom Datasets

```
telcom_all = pd.concat([telcom,telcom2,telcom3],sort = False,ignore_index= True)

telcom_all.head()
#telcom_all.loc['x']
```

```
telcom_all.isnull().sum()
```

```
#merge the telcom_all and cell_geo dataset
#Joined based on index
users = pd.merge(cell_geo,telcom_all,how= 'left' , left_on= 'site_code', right_on='site_id')
users.isnull().sum()
users.shape
users
```

```
# Creating a subset dataframe to use for analysis
users.drop(['decoupzone'	,'zonename','longitude','latitude','area','cell_id_x','dw_a_number_int','dw_b_number_int','country_a','country_b','cell_id_y','cell_on_site'], axis = 1, inplace= True)
users
```

```
users.info()
```

```
# check for duplicated values
users.duplicated(['datetime'])
#drop duplicated values
users_duplicated = users.drop_duplicates()
users_duplicated

#check for null values
users_duplicated.isnull().sum()
#dropping null values
users_duplicated.dropna(how= 'all')
users_drop = users_duplicated.dropna(axis=0, subset=['product','datetime'] )
users_drop
```

```
date = users_drop['datetime'].apply(lambda x : pd.Series([i for i in x.split(" ")]))
date.columns = ['date','time']
date
```

```
users_fin = pd.concat([users_drop,date],axis=1)
users_final = users_fin.drop(['datetime'], axis=1)
users_final
```

```
#how many cities we have
city_no = users_final['villes'].unique()
len(city_no)
```

```
#cities that were used most
#nlargest() returns the largest n elements.
cities = users_final['site_id'].groupby(users_final['villes']).count().sort_values(ascending = False).nlargest(15)
cities
  
```
